```python
from bs4 import BeautifulSoup
import requests
import re
import json
import pandas as pd
```

```python
website = 'https://minfin.com.ua/banks/rating/?date='
```

```python
date = '2016-01-01' # input('Enter date in format YYYY-MM-DD: ')
```

```python
site = (website + date)
print(site)
```

```python
response = requests.get(site)
```

```python
content = response.text
```

```python
soup = BeautifulSoup(content, 'lxml')
```

```python
print(soup.prettify())
```

```python
with open('output.xml', 'w') as file:
    file.write(soup.prettify())
```

```python
div = soup.find('div', class_='clear rating-pag sustain')
```

```python
print(div)
```

```python
data = div.find('script')
```

```python
print(data.text)
```

```python
data = data.text.replace("data =", "")
```

```python
data = data.replace("\'", "\"")
```

```python
print(data)
```

```python

with open("data.txt", "w") as file:
    file.write(data)
```

```python
pattern = re.findall(r"\,[0-9]+:", data)
```

```python
first = re.findall(r"\s[0-9]+:", data)
print (first)   
```

```python
print(pattern)
```

```python

df = pd.read_csv('banks.csv')
print (df)
print (df.dtypes)
df['id'] = df['id'].astype(str)
```

```python
for p in pattern:
    bank_id = p.replace(":", "")
    bank_id = bank_id.replace(",", "")
    print (bank_id)
    bank_name = df.loc[df["id"] == bank_id ].values[0][1]
    print (bank_name)
```

```python


# Convert the redacted data to a dictionary
redacted_dict = json.loads(redacted_data)

# Save the redacted data to a JSON file
with open('redacted_data_'+date+'.json', 'w') as file:
    json.dump(redacted_dict, file)
```

```python


# Convert the redacted_dict to a DataFrame
df = pd.DataFrame.from_dict(redacted_dict)

# Save the DataFrame as a CSV file
df.to_csv('redacted_data_'+date+'.csv', index=False)
```
